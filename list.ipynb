**List**
Lists are used to store multiple items in a single variable.
Lists are one of 4 built-in data types in Python used to store collections of data,the other 3 are 
1. Tuple 
2. Set
3. Dictionary

**Lists are created using square brackets.

------------------------------******------------------------------------------

List Items

List items are 

1.ordered, 

Ordered
When we say that lists are ordered, it means that the items have a defined order, and that order will not change.

If you add new items to a list, the new items will be placed at the end of the list.


------------------------------******------------------------------------------


2.changeable, and allow 
The list is changeable, meaning that we can change, add, and remove items in a list after it has been created.


------------------------------******------------------------------------------


3.duplicate values.


Lists allow duplicate values


List items are indexed, the first item has index [0], the second item has index [1] etc.



```python
#append() method appends an element to the end of the list.

lists = ["apple", "banana", "cherry"]
lists.append("orange")
lists.append("orange")
print("*Append")
print(lists)


#copy() method returns a copy of the specified list.
listCopy=lists.copy()
print("*List Copy")
print(listCopy)


#count() method returns the number of elements with the specified value.
#Return the number of times the value 9 appears int the list:
listCount=lists.count("orange")
print("*List Count Orange")
print(listCount)

#extend() method adds the specified list elements (or any iterable) to the end of the current list.
points = [1, 4, 5, 9]
lists.extend(points)
print("*List Added")
print(lists)


#index() method returns the position at the first occurrence of the specified value.
#index() method only returns the first occurrence of the value.
print("*List Index Number")
indexNumber=lists.index("orange")
print(indexNumber)

#insert() method inserts the specified value at the specified position.
print("*Insert Specific Position")
lists.insert(1, "apple2")
print(lists)

#pop() method removes the element at the specified position.
print("*Pop Specific Position")
lists.pop(1)
print(lists)

#remove() method removes the first occurrence of the element with the specified value.
print("*Remove Specific Value")
lists.remove(5)
print(lists)

#reverse() method reverses the sorting order of the elements.
print("*Print Reverse in List ")
lists.reverse()
print(lists)

#Sort the list alphabetically
print("*Sort in Alphabet List ")
lists2= ['Ford', 'BMW', 'Volvo']
lists2.sort()
print(lists)

#Sort the list Number
print("*Sort in Number List ")
points2 = [1, 4, 3, 9]
points2.sort()
print(points2)

#determine how many items a list has, use the len() function
print("*Lists Length")
print(len(lists))

#What is the data type of a list?
print("*Lists Type")
print(type(lists))


#list() constructor to make a List:
print("*Lists Constructor")
thislist = list(("apple", "banana", "cherry")) # note the double round-brackets
print(thislist)


# loop through the list items by using a for loop:
print("*List Print Using Loop")
for x in lists:
  print(x)


print("*List Print Using list index number and  Loop")
for i in range(len(lists)):
  print(lists[i])
print("*List Print Using while loop")
i = 0
while i < len(lists):
  print(lists[i])
  i = i + 1

#short hand for loop that will print all items in a list
print("*List Print Using short hand for loop")
[print(x) for x in lists]

#List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.
print("*List added new list using existing list")
lists2 = ["apple","apple2","banana", "cherry"]
newlist = []
for x in lists2:
  if "a" in x:
    newlist.append(x)

print(newlist)

#With list comprehension you can do all that with only one line of code:
print("*List added new list using existing list Short Method")
newlist2 = [x for x in lists2 if "a" in x]
print(newlist2)


print("*List added new list using existing list Short Method Only accept items that are not apple:")
newlist3 = [x for x in lists2 if x != "apple"]
print(newlist3)

newlist4 = [x for x in lists2]

#range() function to create an iterable:
newlist5 = [x for x in range(10)]


#clear() method removes all the elements from a list.
print("*Lists Full Clear")
lists.clear()
print(lists)




#stack
#queue
#Tuple
#Sequence
#Sets
#Dictionary
```

```python
class Room:
  def __init__(self, dirt_amount):
    self.dirt_amount = dirt_amount

class VacuumCleaner:
  def clean(self, room):
    print(f"Cleaning room with {room.dirt_amount} units of dirt.")
    room.dirt_amount = 0

# create a room with 10 units of dirt
room = Room(10)

# create a vacuum cleaner
vacuum_cleaner = VacuumCleaner()

# clean the room
vacuum_cleaner.clean(room)

# check if the room is clean
if room.dirt_amount == 0:
  print("Room is now clean.")
```
